## Why These Charts Matter in ML

* Bar Chart → Class distribution, feature importance.
* Pie Chart → Category proportions (less used in ML but sometimes helpful).
* Histogram → Feature distribution (crucial for feature engineering, normalization).
* Scatter Plot → Relationship between two features (used for EDA - Exploratory Data Analysis).

## 1. Bar Chart in Matplotlib

Use in ML

* Show class imbalance.
* Show feature importances.

**Code Example**

```python
# Example: Visualizing class distribution
classes = ['Cat', 'Dog', 'Rabbit']
counts = [45, 30, 25]

plt.figure(figsize=(6, 4))
plt.bar(classes, counts, color='skyblue')
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Practical Note:**

* Use this before training a classifier to check imbalance.
* Use plt.barh() for horizontal bars.

## 2. Pie Chart in Matplotlib

Use in ML

* Not very common in ML, but good for visualizing proportions quickly.

**Code Example**

```python
# Example: Proportions of labels
labels = ['Positive', 'Negative', 'Neutral']
sizes = [50, 30, 20]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Sentiment Analysis Result")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

**Practical Note:**

* Use autopct='%1.1f%%' to show percentage.
* Less preferred in papers—harder to interpret with many categories.

## 3. Histogram in Matplotlib


Use in ML

* Analyze feature distribution (skewed, normal, etc.)
* Check if you need to normalize or transform the feature.

**Code Example**

```python
# Simulate a feature (e.g., age of users)
np.random.seed(0)
ages = np.random.normal(loc=30, scale=5, size=100)

plt.figure(figsize=(6, 4))
plt.hist(ages, bins=10, color='salmon', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Practical Note:**

* Try different bins (e.g., 10, 20) to see distribution clearly.
* Helps you choose the right scaling or transformation (e.g., log, standard scaling).

## 4. Scatter Plot in Matplotlib

Use in ML

* Visualize relationships between two features.
* Check correlation, clusters, and outliers.

**Code Example**

```python
# Simulated ML features (e.g., height vs. weight)
np.random.seed(0)
height = np.random.normal(160, 10, 100)
weight = height * 0.5 + np.random.normal(0, 5, 100)

plt.figure(figsize=(6, 4))
plt.scatter(height, weight, color='teal', alpha=0.7)
plt.title("Height vs. Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)
plt.show()
```

**Practical Note:**

* Use alpha=0.7 for transparency if points overlap.
* Can use color=label to plot different classes in one chart.
* Useful before building models like Linear Regression or Clustering.

## Scatter Plot with Labels (for ML Clustering)

```python
# Simulate 2D features + labels (e.g., k-means clusters)
x = np.random.rand(100)
y = np.random.rand(100)
labels = np.random.randint(0, 3, 100)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, c=labels, cmap='viridis')
plt.title("Cluster Visualization")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label='Cluster ID')
plt.show()
```

## Summary Table

| Chart | Usage in ML | Notes |
| --- | --- | --- |
| Bar Chart | Class count, feature importance | Use plt.bar, easy to compare values |
| Pie Chart | Proportion of categories | Use plt.pie, but not ideal for many labels |
| Histogram | Feature distribution, outliers | Use plt.hist, explore skewness & scale |
| Scatter Plot | Relationship between 2 features | Use plt.scatter, crucial for EDA |
