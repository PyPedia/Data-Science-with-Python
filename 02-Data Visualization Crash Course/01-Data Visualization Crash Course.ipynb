**5. Basic Plotting with Matplotlib**

The first step in visualizing data is creating basic plots using Matplotlib.

***Line Plot:***

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

***Bar Plot:***

```python
plt.bar([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Simple Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

***Scatter Plot:***

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

**6. Customizing Your Plots**
    
You can customize your plots by changing their *styles*, *colors*, and *labels*.

```python
plt.plot(x, y, linestyle='--', marker='o', color='g')  # Custom line style and color
plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

**1. Line Plot with Matplotlib**

Line plots are useful for visualizing data points over time or any continuous variable. They are the most common plot type in data analysis.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, label='y = x^2', color='blue', marker='o')  # Line plot with customizations
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```

**2. Bar Plot with Seaborn**

Bar plots are great for comparing categorical data. Heres how to create a bar plot using Seaborn:

```python
import seaborn as sns
import pandas as pd

data = {'Category': ['A', 'B', 'C', 'D'], 'Value': [10, 20, 15, 25]}
df = pd.DataFrame(data)

sns.barplot(x='Category', y='Value', data=df, palette='Blues')  # Simple bar plot
plt.title('Bar Plot Example')
plt.show()
```

**3. Box Plot for Distribution**

Box plots show the distribution of data and highlight outliers, medians, and quartiles.

```python
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot Example')
plt.show()
```

**4. Heatmap for Correlation**

Heatmaps are great for showing the correlation between different variables in your dataset.

```python
data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
})

corr = data.corr()  # Compute correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**5. Pair Plot for Multivariable Analysis**
    
Pair plots are used to visualize pairwise relationships between variables, great for exploratory data analysis.

```python
iris = sns.load_dataset('iris')  # Loading a sample dataset
sns.pairplot(iris, hue='species')
plt.show()
```

**6. Interactive Plot with Plotly**

Plotly allows you to create interactive visualizations that you can explore by hovering, zooming, and more.

```python
import plotly.express as px

data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 11, 12, 13, 14]
})

fig = px.line(data, x='x', y='y', title='Interactive Line Plot')
fig.show()  # This opens an interactive chart in your web browser
```

**7. 3D Scatter Plot with Plotly**

You can create 3D plots for visualizing relationships between three variables.

```python
fig = px.scatter_3d(data, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```

**8. Histograms with Matplotlib and Seaborn**

Histograms are useful for showing the distribution of a single variable.

```python
data = np.random.randn(1000)  # Generate 1000 random numbers
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram Example')
plt.show()
```

Or using Seaborn:

```python
sns.histplot(data, kde=True, color='red')
plt.title('Seaborn Histogram')
plt.show()
```

**11. Subplots for Multiple Plots**
    
Sometimes you need to display multiple plots in one figure. You can do this using subplots:

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].plot(x, y)
axs[0].set_title('Subplot 1')
axs[1].bar(df['Category'], df['Value'])
axs[1].set_title('Subplot 2')
plt.tight_layout()
plt.show()
```
