## Step 1: What is a Format String?

In Matplotlib, plot() can accept a format string to control the appearance of the plotted line.

**Syntax:**

```python
plt.plot(x, y, 'format_string')
```

The format string is:

```python
'color marker linestyle'
```

Each part is optional, but the order must be correct.

## Step 2: Format String Codes

| Element | Symbol | Example | Meaning |
| --- | --- | --- | --- |
| Color | 'r', 'g', 'b', 'k', 'y', 'c', 'm', 'w' | 'r' | red, green, blue, etc. |
| Marker | 'o', '^', 's', 'x', 'D', '+' | 'o' | circle, triangle, square, etc. |
| Line Style | '-', '--', '-.', ':' | '--' | solid, dashed, dash-dot, dotted |

## Step 3: First Plot Example

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, 'ro--')  # red circles, dashed line
plt.title("Simple Format String Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

**Explanation:**
  
* 'r': red color
* 'o': circle markers
* '--': dashed line

## Step 4: Try More Examples (Practice Time!)

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b-.^', label='sin(x)')  # blue, dash-dot, triangle marker
plt.plot(x, y2, 'g:x', label='cos(x)')  # green, dotted, x marker

plt.legend()
plt.title("Multiple Format Strings")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

## Step 5: Real Machine Learning Example (Loss Curve)

```python
epochs = list(range(1, 11))
loss_train = [0.9, 0.8, 0.7, 0.6, 0.5, 0.42, 0.35, 0.30, 0.28, 0.25]
loss_val = [1.0, 0.95, 0.85, 0.75, 0.67, 0.6, 0.55, 0.52, 0.5, 0.49]

plt.plot(epochs, loss_train, 'bo-', label='Train Loss')  # blue circles with solid line
plt.plot(epochs, loss_val, 'rs--', label='Validation Loss')  # red squares with dashed line

plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

## Why use format strings here?

* They visually differentiate multiple curves
* They make plots easier to interpret
* They save time compared to using color=, marker=, linestyle= separately

## Step 6: Combining With Keyword Arguments (Advanced Tip)

You can combine format strings with explicit arguments for more control:

```python
plt.plot(x, y1, 'go--', linewidth=2, markersize=8, label='Custom Line')
```

This gives you:

* Green, circle marker, dashed
* Thicker line, larger markers

## Step 7: Tips for Machine Learning Use

| Use Case | Tip |
| --- | --- |
| Loss, accuracy plots | Use different marker shapes and line styles for train/val/test |
| Confusion matrix as line | Use solid lines for ground truth, dashed for predictions |
| Learning rate changes | Use ':D' (dotted + diamond marker) for schedules |
