## Step 2: Your First Basic Plot

**Example: Plotting a Line**

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)        # Plot the graph
plt.title("Basic Line Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

**Explanation:**

* x and y: These are lists of coordinates.
* plt.plot(): Plots the graph (by default, a line).
* plt.title(), plt.xlabel(), plt.ylabel(): Add context.
* plt.grid(True): Adds a grid.
* plt.show(): Renders the graph window.

## Step 3: Plot Multiple Lines

**Example:**

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 22]

plt.plot(x, y1, label="Model A")
plt.plot(x, y2, label="Model B")

plt.title("Model Performance")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

**Tip for ML:**

Use this method to compare training vs validation accuracy, or multiple model results.

## Step 4: Add Style

```python
plt.plot(x, y1, color='blue', linestyle='--', marker='o', label="Training")
plt.plot(x, y2, color='green', linestyle='-', marker='x', label="Validation")
```

| Parameter | Usage |
| --- | --- |
| color | 'red', 'blue', '#00FF00' |
| linestyle | '-', '--', '-.', ':' |
| marker | 'o', 's', '^', 'x', 'D' |

## Step 7: Bar Charts (Useful for Feature Importance)

```python
features = ['Feature A', 'Feature B', 'Feature C']
importance = [0.3, 0.5, 0.2]

plt.bar(features, importance, color='skyblue')
plt.title("Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.show()
```

## Step 8: Scatter Plots (Good for Visualizing Clusters)

```python
x = [1, 2, 3, 4]
y = [10, 15, 13, 18]

plt.scatter(x, y, color='purple', marker='x')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

Use this when visualizing data distribution, k-means clusters, etc.

## Step 9: Subplots (Multiple Plots in One Figure)

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # (rows, cols, index)
plt.plot(x, y1)
plt.title("Sine Function")

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Cosine Function")

plt.tight_layout()
plt.show()
```

Subplots are ideal when you want to compare functions or results visually.

## Step 10: Plotting From Pandas (For ML Workflows)

```python
import pandas as pd

data = {
    'epoch': [1, 2, 3, 4],
    'train_accuracy': [0.6, 0.7, 0.8, 0.85],
    'val_accuracy': [0.58, 0.68, 0.75, 0.8]
}

df = pd.DataFrame(data)

plt.plot(df['epoch'], df['train_accuracy'], label='Training')
plt.plot(df['epoch'], df['val_accuracy'], label='Validation')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

Real machine learning projects are usually built on Pandas DataFrames.
