## Goal

In machine learning, plotting multiple lines, loss curves, accuracy metrics, etc., requires:
* Proper **labels** for lines/plots.
* A **legend** to identify which line represents what.

## Step 1: Setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Step 2: Create Dummy Data (Simulate ML Metrics)

Let's pretend we're plotting training and validation loss from a neural network.

```python
epochs = np.arange(1, 21)  # 1 to 20
train_loss = np.exp(-0.2 * epochs) + 0.1 * np.random.randn(20)
val_loss = np.exp(-0.2 * epochs) + 0.15 * np.random.randn(20)
```

## Step 3: Plot with Labels

```python
plt.plot(epochs, train_loss, label='Training Loss', color='blue', linestyle='--', marker='o')
```

```python
plt.plot(epochs, val_loss, label='Validation Loss', color='red', linestyle='-', marker='x')
```

**Explanation of Parameters:**

| Parameter | What it does | Practical Tip (ML Context) |
| --- | --- | --- |
| label | Used by legend() | Always label what each line means (e.g., accuracy, loss, etc.) |
| color | Line color | Use consistent color schemes (e.g., blue for training) |
| linestyle | Solid (-), dashed (--), etc. | Helps visually differentiate curves |
| marker | Adds symbols to data points | Useful to highlight actual epochs/points |

## Step 4: Add Titles and Axis Labels

```python
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

**Tip:**
  
Always include titles and axis labels in ML plots — makes your results clearer in reports or papers.

## Step 5: Add Legend

```python
plt.plot(epochs, train_loss, label='Training Loss', color='blue', linestyle='--', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', color='red', linestyle='-', marker='x')

plt.legend()
```

**Legend Customization:**

```python
plt.plot(epochs, train_loss, label='Training Loss', color='blue', linestyle='--', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', color='red', linestyle='-', marker='x')

plt.legend(loc='upper right', fontsize='large', shadow=True, fancybox=True)
```

| Parameter | Description |
| --- | --- |
| loc | Position: 'upper right', 'lower left', 'center', etc. |
| fontsize | 'small', 'medium', 'large', or numeric value |
| shadow | Adds a shadow behind the legend |
| fancybox | Rounded corners for the legend box |

## Full Example Code

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate ML loss data
epochs = np.arange(1, 21)
train_loss = np.exp(-0.2 * epochs) + 0.1 * np.random.randn(20)
val_loss = np.exp(-0.2 * epochs) + 0.15 * np.random.randn(20)

# Plot
plt.plot(epochs, train_loss, label='Training Loss', color='blue', linestyle='--', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', color='red', linestyle='-', marker='x')

# Labels and title
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Legend
plt.legend(loc='upper right', fontsize='medium', shadow=True, fancybox=True)

# Final touches
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Use in Machine Learning Projects

Plotting Model Metrics:

* **Loss** and **accuracy** for training/validation.
* **Precision**, **recall**, **F1-score** trends.
* Compare **multiple models** on the same plot (e.g., Logistic Regression vs. Random Forest).

## Plot Multiple Models

```python
# Simulated accuracy data
model1 = 0.6 + 0.02 * epochs + 0.02 * np.random.randn(20)
model2 = 0.55 + 0.025 * epochs + 0.01 * np.random.randn(20)

plt.plot(epochs, model1, label='Model A Accuracy', color='green')
plt.plot(epochs, model2, label='Model B Accuracy', color='purple')

plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
